```python
# Scatter Plot: A scatter plot is a type of chart that displays data as a collection of points. Each point represents the value of two variables.
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
plt.scatter(x,y)
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.title('Scatter Plot Diagram')
```

```python
import matplotlib.pyplot as plt
import numpy as np
fig, axs = plt.subplots(2, 2)
# Create the data for the plots
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])
axs[0, 0].plot(x1, y1)
axs[0, 0].set_title('Line 1')

x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])
axs[0, 1].plot(x2, y2)
axs[0, 1].set_title('Line 2')

x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

axs[1, 0].plot(x3, y3)
axs[1, 0].set_title('Line 3')


x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

axs[1, 1].plot(x4, y4)
axs[1, 1].set_title('Line 4')


# Add a title for the overall figure
fig.suptitle('Four Line Plots')

# Display the plots
plt.show()
```

```python
# Bar Plot: A bar plot is a type of chart that displays data as rectangular bars. The length of each bar represents the value of a variable.
import matplotlib.pyplot as plt
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

plt.xlabel('company')
plt.ylabel('profit')
plt.title('Data of Companies')
plt.bar(company,profit)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

plt.xlabel('company')
plt.ylabel('profit')
plt.title('Data of Companies')
plt.barh(company,profit)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
data = [box1,box2]
fig,ax = plt.subplots()
ax.boxplot(data)
ax.set_title('Box Plot')
ax.set_xlabel('BOX1')
ax.set_ylabel('BOX2')
plt.show()
```
