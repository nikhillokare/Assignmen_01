```python
# Estimation statistics is a branch of statistics that deals with estimating population parameters based on sample statistics.
# Point estimate refers to a single value that is used to estimate the population parameter, such as the sample mean used to estimate the population mean. 
# Interval estimate refers to a range of values that is likely to contain the population parameter, such as a confidence interval.
```

```python
def estimate_population_mean(sample_mean, sample_std, sample_size):
    # Calculate the standard error of the mean
    std_error = sample_std / (sample_size ** 0.5)
    
    # Calculate the lower and upper bounds of the 95% confidence interval
    lower = sample_mean - 1.96 * std_error
    upper = sample_mean + 1.96 * std_error
    
    return (lower, upper)
```

```python
# Hypothesis testing is a statistical method used to make decisions about a population based on sample data. 
# It involves making a hypothesis about the population parameter, collecting sample data, and then making a decision about whether to reject or fail to reject the hypothesis. 
# Hypothesis testing is used to determine whether the observed differences between groups are due to chance or whether they are statistically significant.
```

```python
# Hypothesis: The average weight of male college students is greater than the average weight of female college students.
```

```python
import numpy as np
from scipy import stats

# Sample data
male_weights = [180, 185, 190, 175, 195]
female_weights = [150, 160, 155, 165, 170]

# Calculate the sample means and standard deviations
male_mean = np.mean(male_weights)
female_mean = np.mean(female_weights)
male_std = np.std(male_weights, ddof=1)
female_std = np.std(female_weights, ddof=1)

# Conduct a two-sample t-test assuming unequal variances
t, p = stats.ttest_ind_from_stats(male_mean, male_std, len(male_weights),
                                  female_mean, female_std, len(female_weights),
                                  equal_var=False)

# Print the results
print("t-value: {:.2f}".format(t))
print("p-value: {:.4f}".format(p))
```

```python
# A null hypothesis is a statement that there is no significant difference between two populations or that a population parameter is equal to a specific value. 
# An alternative hypothesis is a statement that there is a significant difference between two populations or that a population parameter is not equal to a specific value.

# Examples of null hypotheses:

# The average height of men and women is the same.
# There is no difference in test scores between students who studied with a tutor and those who did not.
# Examples of alternative hypotheses:

# The average height of men is greater than the average height of women.
# Students who studied with a tutor scored higher on the test than those who did not.
```

```python
# State the null and alternative hypotheses.
# Choose the level of significance (alpha) and determine the critical value or p-value.
# Collect sample data and calculate the test statistic.
# Compare the test statistic to the critical value or p-value.
# Make a decision to reject or fail to reject the null hypothesis.
# Interpret the results and draw conclusions.
```

```python
# The p-value is the probability of obtaining a test statistic as extreme or more extreme than the observed value, assuming the null hypothesis is true.
# A small p-value (typically less than 0.05) suggests that the null hypothesis should be rejected, while a large p-value suggests that there 
# is not enough evidence to reject the null hypothesis. The p-value is significant in hypothesis testing because it provides a quantitative measure of the strength
# of the evidence against the null hypothesis.
```

```python
from scipy.stats import ttest_ind

sample1 = [5, 7, 9, 11, 13]
sample2 = [4, 6, 8, 10, 12]
alpha = 0.05

t_statistic, p_value = ttest_ind(sample1, sample2)
print("t-statistic:", t_statistic)
print("p-value:", p_value)

if p_value < alpha:
    print("Reject null hypothesis, population means are not equal.")
else:
    print("Fail to reject null hypothesis, population means are equal.")
```

```python
# t-statistic is a test statistic used in hypothesis testing to determine whether a sample mean is significantly different from a hypothesized population mean. The formula for the t-statistic is:

# t = (x̄ - μ) / (s / √n)

# where:
# x̄ = sample mean
# μ = hypothesized population mean
# s = sample standard deviation
# n = sample size
```

```python
import scipy.stats as stats

n = 50
x_bar = 500
s = 50
alpha = 0.05

t_value = stats.t.ppf(1 - alpha/2, n-1)
lower = x_bar - (t_value * s / np.sqrt(n))
upper = x_bar + (t_value * s / np.sqrt(n))

print("95% Confidence Interval for Population Mean Revenue:")
print("Lower bound: $", round(lower, 2))
print("Upper bound: $", round(upper, 2))
```
