```python
Q5.
Inheritance is a mechanism in Object-Oriented Programming (OOP) that allows a new class to be derived from an existing class. 
The derived class inherits the attributes and behavior of the base class. 
The derived class is also called a subclass or child class, and the base class is also called a superclass or parent class.

1.Single Inheritance:
    class Animal:
    def __init__(self, name):
        self.name = name
        
    def make_sound(self):
        print("Animal sound")
        
class Dog(Animal):
    def __init__(self, name, breed):
        Animal.__init__(self, name)
        self.breed = breed
        
    def make_sound(self):
        print("Bark")
        
dog = Dog("Rover", "Labrador")
dog.make_sound()

2.Multi-level Inheritance:
    class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age
        
    def introduce(self):
        print(f"Hi, my name is {self.name} and I'm {self.age} years old.")
        
class Student(Person):
    def __init__(self, name, age, enrollment_number):
        Person.__init__(self, name, age)
        self.enrollment_number = enrollment_number
        
class GraduateStudent(Student):
    def __init__(self, name, age, enrollment_number, thesis_topic):
        Student.__init__(self, name, age, enrollment_number)
        self.thesis_topic = thesis_topic
        
grad_student = GraduateStudent("Jane", 25, "GS001", "Machine Learning")
grad_student.introduce()


3.Multiple Inheritance:
    class Parent1:
    def __init__(self, name):
        self.name = name

    def introduce(self):
        return f"Hi, I'm {self.name}."

class Parent2:
    def __init__(self, age):
        self.age = age

    def introduce(self):
        return f"I'm {self.age} years old."

class Child(Parent1, Parent2):
    def __init__(self, name, age):
        Parent1.__init__(self, name)
        Parent2.__init__(self, age)

child = Child("John", 32)
print(child.introduce())

Hierarchical Inheritance:
    class Animal:
    def __init__(self, name):
        self.name = name

    def make_sound(self):
        pass

class Dog(Animal):
    def make_sound(self):
        return "Woof woof"

class Cat(Animal):
    def make_sound(self):
        return "Meow meow"

dog = Dog("Max")
cat = Cat("Mittens")

print(dog.name + " says " + dog.make_sound())
print(cat.name + " says " + cat.make_sound())

Hybrid Inheritance:
    class Animal:
    def __init__(self, name):
        self.name = name

class Mammal(Animal):
    def __init__(self, name, breed):
        super().__init__(name)
        self.breed = breed

class Carnivore(Mammal):
    def __init__(self, name, breed, prey):
        super().__init__(name, breed)
        self.prey = prey

class Lion(Carnivore, Animal):
    def __init__(self, name, breed, prey, roar):
        Carnivore.__init__(self, name, breed, prey)
        self.roar = roar

lion = Lion("Simba", "African Lion", "Gazelle", "Roar!")
print("Name:", lion.name)
print("Breed:", lion.breed)
print("Prey:", lion.prey)
print("Roar:", lion.roar)
```
